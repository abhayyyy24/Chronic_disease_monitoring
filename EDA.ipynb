```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('cleaned-data.csv')
```

```python
print(df.head(10))
```

```python
features=['Heart Rate (BPM)','Blood Oxygen Level (%)', 'Step Count','Sleep Duration (hours)','Activity Level','Stress Level','Activity Score' ]
```

```python
for col in features:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'{col} Distribution')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

```python
for col in features:
    plt.figure(figsize=(4,5))
    sns.boxplot(y=df[col])
    plt.title(f'{col} Box Plot')
    plt.tight_layout()
    plt.show()
```
